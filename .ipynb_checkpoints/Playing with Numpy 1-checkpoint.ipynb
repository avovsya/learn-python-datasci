```python
import numpy as np
np_height = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_weight = np.array([65.4, 59.2, 63.6, 88.4, 68.7])

bmi = np_weight / np_height ** 2

print(bmi.round(3))

print(bmi > 23)

print(bmi[bmi > 21.7].round(3)) # WTF?? How this even works???
```

```python
# 2D arrays

# np2d = np.array([[1.73, 1.68, 1.71, 1.89, 1.79], [65.4, 59.2, 63.6, 88.4, 68.7]])
np2d = np.array([np_height, np_weight])
np2dstr = np.array([[1.73, 1.68, 1.71, 1.89, 1.79], [65.4, 59.2, 63.6, 88.4, "68.7"]]) # All elements coverted to strings

print("Shape:", np2d.shape) # Get the shape of the array (2 rows, 5 columns)

print(np2d)

print("Stringified:", np2dstr)

np2d[0][0] = "42" # Will convert string to float
print(np2d)


np2d[0][0] = "aaa" # Won't covert all other elements to strings
print(np2d)
```

```python
# Subsetting multiD arrays

print(np2d)

print(np2d[0][0]) # [row][col]

print(np2d[0, 0]) # [row, col]

print(np2d[:, 0:2]) # [rowStart:rowEndExclusive, colStart:colEndExclusive]
```

```python
print(np2d * 2)

print(np2d[0, :] * 2)

print(np2d[:, 0] * 2)
```

```python
# Math ops on 2d arrays

np_mat = np.array([[1, 2],
                   [3, 4],
                   [5, 6]])
print(np_mat * 2)

print(np_mat + np.array([10, 10]))

print(np_mat + np_mat)
```
